(c) Exponential distribution: Exp(0; 1=1:5). Plot interval: [−1; 8]
(d) Gamma distribution: Gamma(5; 0; 1). Plot interval: [0; 15]
(e) Beta distribution: Beta(2; 1:5; 0; 1). Plot interval: [0; 1]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta,expon,gamma,laplace,norm
```

```python
distributions=[norm(5,1.5),laplace(5,1.5),expon(0,1/1.5),gamma(5,0,1),beta(2,1.5,0,1)]
intervals=[(0,10),(-5,15),(-1,8),(0,15),(0,1)] 
labels=['Normal','Laplace','Exponential','Gamma','Beta']
```

```python
for distribution,interval,label in zip(distributions,intervals,labels):
    plt.figure(figsize=(9, 4))
    plt.subplot(121)
    plt.hist(distribution.rvs(1000), bins=25)
    plt.title(label + ' Historgram')
    plt.subplot(122)
    x = np.linspace(interval[0], interval[1])
    plt.plot(x, distribution.pdf(x))
    plt.title(label + ' PDF')
```

Alternative approach by defining a function to plot distribution


```python
def plot_distribution(distribution, x_min, x_max, sample_size=1000):
    _, ax = plt.subplots(2, 1)
    x = np.linspace(x_min, x_max)
    ax[0].plot(x, distribution.pdf(x))
    ax[1].hist(distribution.rvs(sample_size))
    plt.show()
```

```python
plot_distribution(norm(5,1.5),0,10)
```
